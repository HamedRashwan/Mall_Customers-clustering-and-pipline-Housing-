```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans , AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
```

```python
x=df[['Annual Income (k$)','Spending Score (1-100)']]
x
```

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
x_scaled
```

```python
model=KMeans(n_clusters=3)
model.fit(x_scaled)
```

```python
clusters=model.predict(x_scaled)
```

```python
clusters
```

```python
x
```

```python
x['clusters']=clusters
x
```

```python
centroids=model.cluster_centers_
centroids
```

```python
sns.scatterplot(x=x_scaled[:,0],y=x_scaled[:,1],c=x['clusters'])
plt.scatter(centroids[:,0],centroids[:,1],s=30,c='red')
```

```python
model.inertia_
```

```python
#ELBOW METHOD
scores=[]
for i in range(1,11):
    model=KMeans(n_clusters=i)
    model.fit(x_scaled)
    scores.append(model.inertia_)
    

plt.plot(range(1,11),scores)
plt.xlabel('n_clusters')
plt.ylabel('inertia')
```

```python
df=pd.read_csv('D:\COURSES\AI\Epsilon Ai\Session 35\Train.csv')
df
```

```python
df.drop('ID',axis=1,inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True,ignore_index=True)
```

```python
df.dropna(axis=0,inplace=True)
```

```python
df.reset_index(inplace=True,drop=True)
df
```

```python
cat_col=df.select_dtypes('object_').columns
cat_col
```

```python
for col in df.select_dtypes('object_').columns:
    display(df[col].value_counts().to_frame().T)
```

```python
indexes=detect_outliers(df,0,df.select_dtypes('number').columns)
len(indexes)
```

```python
df.drop(indexes,axis=0,inplace=True)
```

```python
df
```

```python
cat_col
```

```python
encoder=ColumnTransformer(transformers=[('ohe',OneHotEncoder(sparse=False,drop='first'),cat_col)],
                         remainder='passthrough')
encoder
```

```python
pipeline=make_pipeline(encoder,
                                             StandardScaler(),
                                           DBSCAN(eps=1.5,min_samples=70))
pipeline
```

```python
labels=pipeline.fit_predict(df)
```

```python
df['labels']=labels
```
